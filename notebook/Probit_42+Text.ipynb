### Probit Regression  ###

```python
class ProbitModel:
    def __init__(self, X_train, y_train, X_test, y_test):
        self.X_train = X_train
        self.y_train = y_train
        self.X_test = X_test
        self.y_test = y_test

    def fit_model(self, alpha=0.2):  # Adjust alpha to increase regularization strength
        probit_model = sm.Probit(self.y_train, self.X_train)
        self.probit_result = probit_model.fit_regularized(alpha=alpha)

    def predict(self):
        y_pred_proba = self.probit_result.predict(self.X_test)
        y_pred = (y_pred_proba > 0.5).astype(int)
        return y_pred

    def evaluate(self, y_pred):
        auc_prob = roc_auc_score(self.y_test, y_pred)
        print("AUC:", auc_prob)
```
